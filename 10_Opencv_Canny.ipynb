# 使用不同的threshold值來看`Stephen Chow.jpg`的檢測結果

```
o=cv2.imread("Stephen Chow.jpg",cv2.IMREAD_GRAYSCALE) #使用不同的threshold值來看Stephen Chow.jpg的檢測結果
r1=cv2.Canny(o,128,200)
r2=cv2.Canny(o,32,128)
cv2_imshow(o)
cv2_imshow(r1)
cv2_imshow(r2)
```
